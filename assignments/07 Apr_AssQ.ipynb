`Q1. What is the relationship between polynomial functions and kernel functions in machine learning
algorithms?`

`Polynomial functions` and kernel functions are related in machine learning algorithms through the use of kernel methods. Kernel methods are a class of algorithms that operate by implicitly mapping the input data into a high-dimensional feature space using a kernel function, which is often a polynomial function.

In particular, a polynomial kernel function is a type of kernel function that computes the dot product between the feature vectors in a high-dimensional space, without actually computing the mapping itself. This makes it possible to use kernel methods with polynomial functions without explicitly computing the high-dimensional feature space, which can be computationally expensive or even impossible in some cases.

Polynomial kernel functions are commonly used in support vector machines (SVMs) and other kernel-based machine learning algorithms. They can be used to model complex decision boundaries that are not linearly separable in the original feature space. The degree of the polynomial kernel determines the complexity of the decision boundary that can be learned by the model, with higher degrees leading to more complex decision boundaries.

Overall, polynomial functions and kernel functions are related in machine learning through the use of kernel methods, which allow for the implicit mapping of data into a high-dimensional feature space using polynomial functions, without the need to compute the mapping itself.

`Q2. How can we implement an SVM with a polynomial kernel in Python using Scikit-learn?`

To implement an SVM with a polynomial kernel in Python using Scikit-learn, you can use the svm.SVC class and set the kernel parameter to 'poly'. You will also need to set the degree parameter to specify the degree of the polynomial kernel. Here's an example:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm

# Load the iris dataset
iris = datasets.load_iris()

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3,random_state=10)

# Train an SVM with a polynomial kernel
svm_model = svm.SVC(kernel='poly', degree=3)
svm_model.fit(X_train, y_train)

# Test the model
accuracy = svm_model.score(X_test, y_test)
print('Accuracy:', accuracy)
```

In this example, we load the iris dataset and split it into training and test sets. We then train an SVM with a polynomial kernel using the svm.SVC class with kernel='poly' and degree=3. Finally, we test the model on the test set and print the accuracy.

You can adjust the value of degree to change the complexity of the polynomial kernel. Higher values of degree can lead to more complex decision boundaries, but can also increase the risk of overfitting.

`Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?`

In support vector regression (SVR), the parameter `epsilon` controls the width of the margin around the regression line within which errors are tolerated. Increasing the value of `epsilon` allows for larger errors to be tolerated, which results in a wider margin and fewer support vectors.

When `epsilon` is small, the margin around the regression line is narrow, which means that the model will have to fit the data more closely to avoid violating the margin constraint. This can lead to a larger number of support vectors.

On the other hand, when `epsilon` is large, the margin around the regression line is wider, which means that the model can tolerate larger errors without violating the margin constraint. This can result in a smaller number of support vectors, since the model does not need to fit the data as closely.

However, it's important to note that the optimal value of `epsilon` depends on the specific problem and the characteristics of the data. Increasing `epsilon` too much can lead to a decrease in model accuracy, as the model may not be able to capture important patterns in the data. Therefore, it's important to tune the value of `epsilon` carefully to achieve the best performance for a given problem.

`Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter
affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works
and provide examples of when you might want to increase or decrease its value?`

The choice of kernel function, C parameter, epsilon parameter, and gamma parameter can have a significant impact on the performance of Support Vector Regression (SVR). Here's how each parameter works and examples of when you might want to increase or decrease its value:

1. **Kernel Function**: The kernel function is used to transform the input data into a higher-dimensional feature space where it may be more easily separable. The choice of kernel function can have a big impact on the performance of the SVR model.
    * *Linear kernel*: The linear kernel is the simplest kernel function and works well when the data is linearly separable. It has fewer hyperparameters to tune, making it less prone to overfitting.
    * *Polynomial kernel*: The polynomial kernel can capture non-linear relationships between the features. The degree of the polynomial can be adjusted to control the complexity of the model.
    * *RBF kernel*: The RBF (radial basis function) kernel is a popular choice and is good at capturing complex relationships between the features. The gamma parameter controls the width of the Gaussian kernel and can be used to control the complexity of the model.
2. **C parameter**: The C parameter controls the tradeoff between minimizing the error and maximizing the margin. A smaller value of C will lead to a wider margin and more errors tolerated, while a larger value of C will lead to a narrower margin and fewer errors tolerated.
    * *Increase C when*: you have a high degree of confidence in the data and want a more accurate model, or when the data is well separated and you want a narrow margin.
    * *Decrease C when*: you want a wider margin and are willing to tolerate more errors, or when the data is noisy and you want to avoid overfitting.
3. **Epsilon parameter**: The epsilon parameter controls the width of the margin around the regression line within which errors are tolerated. A larger value of epsilon allows for larger errors to be tolerated and results in a wider margin.
    * *Increase epsilon when*: you want a wider margin and are willing to tolerate larger errors, or when the data has a lot of noise.
    * *Decrease epsilon when*: you want a narrower margin and are willing to tolerate fewer errors, or when the data is well-behaved and has a low level of noise.
4. **Gamma parameter**: The gamma parameter controls the width of the Gaussian kernel used in the RBF kernel function. A smaller value of gamma results in a wider kernel and smoother decision boundaries, while a larger value of gamma results in a narrower kernel and more complex decision boundaries.
    * *Increase gamma when*: you want a more complex model that can capture finer details in the data, or when the data is well-behaved and has a low level of noise.
    * *Decrease gamma when*: you want a simpler model that is less prone to overfitting, or when the data has a high level of noise.
It's important to note that the optimal values of these parameters depend on the specific problem and the characteristics of the data. Therefore, it's important to experiment with different values and use techniques such as cross-validation to find the best values for a given problem.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#using the algerian forest fire dataset
df = pd.read_csv('D:/pwskills-Data_Science_Masters/machine learning/ml video practice notebooks/Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.Classes = df.Classes.str.strip()
df["Classes"] = np.where(df.Classes == "fire",1,0)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.drop(['day','month','year'],axis=1).corr().round(2),annot=True)
```

#### INSIGHT
Need to drop BUI to avoid multicollinearity

```python
## Spltting into dependent and independent features
X = df.drop(['day','month','year','Classes',"BUI"],axis=1)
y = df.Classes
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
##Preprocessing the data using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
## SVC
from sklearn.svm import SVC
classifier = SVC(kernel="linear")
classifier.fit(X_train_scaled,y_train)
```

```python
y_pred = classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report , accuracy_score , confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### Hyperparameter tuning with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
# defining parameter ranges
param_grid = {
    "C" : [0.1,1,10,100],
    "gamma" : [1,0.1,0.001,0.0001],
    "kernel" : ["linear",'poly', 'rbf', 'sigmoid']
}
```

```python
classifier = SVC()
clf = GridSearchCV(classifier,param_grid=param_grid,scoring="accuracy",cv=10,verbose=3)
```

```python
clf.fit(X_train_scaled,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

### With hyperparameter tuning the model accuracy has been increased from 93.44% to 97.8%

```python
## classifer with hyperparameters tuned - final model
classifier = SVC(C = 1, gamma = 1, kernel ='linear')
```

```python
classifier.fit(X_train_scaled,y_train)
```
