`Q1. What is the mathematical formula for a linear SVM?`

he mathematical formula for a linear Support Vector Machine (SVM) can be expressed as follows:

Given a training dataset {(x1, y1), (x2, y2), ..., (xn, yn)}, where xi is a d-dimensional input vector and yi is the corresponding binary class label (+1 or -1), the linear SVM seeks to find a hyperplane that separates the positive examples from the negative examples with the largest margin.

The hyperplane can be represented by the equation w · x + b = 0, where w is the weight vector perpendicular to the hyperplane and b is the bias term. The decision function of the SVM is then given by:

f(x) = sign(w · x + b)

where sign is the sign function, which returns +1 if the argument is positive, -1 if it is negative, and 0 if it is zero.

The goal of the SVM is to find the values of w and b that minimize the following objective function:

minimize 1/2 ||w||^2 subject to yi(w · xi + b) ≥ 1 for all i

where ||w|| is the Euclidean norm of the weight vector w. The constraint yi(w · xi + b) ≥ 1 ensures that all training examples are correctly classified and that they lie on the correct side of the margin.

This is a convex optimization problem that can be solved efficiently using techniques such as gradient descent or the quadratic programming algorithm. Once the optimal values of w and b are found, the decision function of the SVM can be used to predict the class label of new examples.

`Q2. What is the objective function of a linear SVM?`

The objective function of a linear Support Vector Machine (SVM) is a convex optimization problem that seeks to find the values of the weight vector w and the bias term b that minimize the following expression:

minimize 1/2 ||w||^2 subject to yi(w · xi + b) ≥ 1 for all i

where xi is a d-dimensional input vector, yi is the corresponding binary class label (+1 or -1), and ||w|| is the Euclidean norm of the weight vector w.

The first term in the objective function, 1/2 ||w||^2, is a regularization term that encourages the weight vector to have small values, thus promoting a simpler and more robust model. The regularization parameter C can be used to control the trade-off between maximizing the margin and minimizing the classification error. A larger value of C results in a smaller margin but fewer training errors, while a smaller value of C leads to a larger margin but more training errors.

The second term in the objective function, yi(w · xi + b), represents the constraint that all training examples must lie on the correct side of the decision boundary (w · x + b = 0). The term yi(w · xi + b) ≥ 1 ensures that the margin between the decision boundary and the closest examples of each class is at least 1, which maximizes the margin and promotes good generalization to unseen data.

The linear SVM objective function is convex, meaning that it has a unique global minimum, and can be solved efficiently using techniques such as gradient descent or the quadratic programming algorithm.

`Q3. What is the kernel trick in SVM?`

The kernel trick is a powerful technique used in Support Vector Machines (SVMs) to enable nonlinear classification by implicitly mapping the input data into a high-dimensional feature space, without actually computing the feature vectors explicitly.

The basic idea behind the kernel trick is to replace the dot product between the input vectors with a nonlinear function that corresponds to a similarity measure between the feature vectors in the high-dimensional space. Specifically, the kernel function K(xi, xj) computes the similarity between the ith and jth input vectors, and is defined as:

K(xi, xj) = φ(xi) · φ(xj)

where φ is a mapping function that transforms the input vectors xi and xj into a high-dimensional feature space, and · represents the dot product between the feature vectors.

Using the kernel trick, the SVM can find the optimal separating hyperplane in the high-dimensional feature space, without explicitly computing the feature vectors. This is computationally efficient and can handle very large input spaces. Furthermore, the kernel trick can enable the SVM to capture complex nonlinear relationships between the input variables, which may not be possible with a linear classifier.

There are several commonly used kernel functions, such as the polynomial kernel, the radial basis function (RBF) kernel, and the sigmoid kernel. The choice of kernel depends on the specific problem at hand and the characteristics of the data.

 `Q4. What is the role of support vectors in SVM Explain with example`

In Support Vector Machines (SVMs), support vectors are the data points that lie closest to the decision boundary (or hyperplane) and have the largest influence on the location and orientation of the boundary.

Support vectors are important because they determine the maximum margin that can be achieved by the SVM and the robustness of the model to variations in the data. The margin is defined as the distance between the decision boundary and the closest data points of each class, and maximizing the margin ensures good generalization to unseen data.

For example, consider a binary classification problem where the input data consists of two features, x1 and x2, and the goal is to separate the data points into two classes, labeled as +1 (blue box) and -1 (green dot), using a linear SVM. The figure below shows a possible decision boundary (represented by a dashed line) that separates the two classes:



In this example, there  data points (shown as Support vectors) that lie closest to the decision boundary and have the largest influence on its position and orientation. These points are the support vectors of the SVM.

The support vectors are determined by the training algorithm of the SVM, which seeks to find the hyperplane that maximizes the margin while ensuring that all training examples are correctly classified. The support vectors lie on the margin and satisfy the following conditions:

The distance from the support vectors to the decision boundary is exactly 1.
The product of the class label and the distance from the decision boundary to any other data point of the same class is greater than or equal to 1.
In this example, the support vectors are the points (1, 1), (-1, -1), and (-1, 1), which correspond to the closest examples of each class to the decision boundary.

The role of the support vectors in the SVM is to define the decision boundary and the margin, and to ensure that the model is robust to small perturbations in the data. Removing a support vector from the training set can affect the position and orientation of the decision boundary, and may lead to poorer generalization performance.
 

`Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?`

Sure, here are some examples and graphs to illustrate the concepts of Hyperplane, Marginal Plane, Soft Margin, and Hard Margin in SVM:

Hyperplane: In SVM, a hyperplane is a decision boundary that separates the data points of two classes in a feature space. In a binary classification problem, the hyperplane is a line in two-dimensional space, a plane in three-dimensional space, and a higher-dimensional hyperplane in higher-dimensional spaces. The hyperplane is defined by the SVM algorithm and is chosen to maximize the margin between the two classes. 


Marginal Plane: In SVM, the marginal plane is a parallel hyperplane that lies at a distance of one margin from the decision boundary (or hyperplane). The marginal plane is important because it defines the boundary beyond which no data points are allowed in a hard-margin SVM, and controls the trade-off between the margin and the classification error in a soft-margin SVM. Here's an example of a hyperplane and marginal plane in a two-dimensional feature space:


!

In this example, the hyperplane is represented by the black line and separates the blue and black data points. The distance between the hyperplane and the closest data points of each class is the margin, which is shown as the distance between the parallel dashed lines.The marginal plane is represented by the dashed lines parallel to the hyperplane (black line), and lies at a distance of one margin from the hyperplane. Any data point that lies between the hyperplane and the marginal plane is considered to be in the margin.

Soft Margin: In SVM, a soft margin is a margin that allows some misclassifications (or errors) in the training data in order to achieve a better generalization performance. Soft margin SVMs are used when the data is not linearly separable or when the margin is too narrow to separate the data without errors. In a soft margin SVM, a penalty parameter C is used to control the trade-off between the margin and the classification error. A large value of C gives more weight to the margin, while a small value of C allows more errors in the training data. 

Hard Margin: In SVM, a hard margin is a margin that does not allow any misclassifications in the training data. Hard margin SVMs are used when the data is linearly separable and the margin is wide enough to separate the data without errors. In a hard margin SVM, the margin is maximized subject to the constraint that all data points are correctly classified. `Q6. SVM Implementation through Iris dataset.`
* Load the iris dataset from the scikit-learn library and split it into a training set and a testing set.
* Train a linear SVM classifier on the training set and predict the labels for the testing set.
* Compute the accuracy of the model on the testing set.
* Plot the decision boundaries of the trained model using two of the features.
* Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use("ggplot")
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
print(dataset.DESCR)
```

```python
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
df["Target"] = dataset.target
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train.shape ,y_train.shape
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel="linear")
classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test)
print("Accuracy is : ",accuracy_score(y_test,y_pred))
```

```python
#Plotting the decision boundaries of the trained model using two of the features.
from sklearn import datasets
from sklearn import svm
# Load iris dataset
iris = datasets.load_iris()

# Select two features
X = iris.data[:, :2]
y = iris.target

# Train SVM model
model = svm.SVC(kernel='linear')
model.fit(X, y)

# Create a grid of points
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict class labels for each point in the grid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("Sepal length vs Sepal width SVM classification of IRIS dataset")
plt.show()
```

```python
#Trying different values of the regularisation parameter C and see how it affects the performance of the model.
from sklearn.model_selection import GridSearchCV
# defining parameter ranges
param_grid = {
    "C" : [0.1,1,10,100] }
grid = GridSearchCV(classifier,param_grid=param_grid,refit=True,cv=5,scoring="accuracy",verbose = 3)
grid.fit(X_train,y_train)
```

```python
#best estimator
grid.best_estimator_
```

```python
#cv results
grid.cv_results_["mean_test_score"]
```

#### As C value is increased the accuracy drops and optimal C value is 1
