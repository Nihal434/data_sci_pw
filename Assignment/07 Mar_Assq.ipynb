Measures of central tendency are summary statistics that represent the center point or typical value of a dataset.The 3 most common measures of central tendency are the mean, median and mode. The mode is the most frequent value. The median is the middle number in an ordered data set. The mean is the sum of all values divided by the total number of values.

The `mean` is the arithmetic average. Calculating the mean is very simple. You just add up all of the values and divide by the number of observations in your dataset.

To find the `median`, order your data from smallest to largest, and then find the data point that has an equal number of values above it and below it. `Covariance` is a statistical term that refers to a systematic relationship between two random variables in which a change in the other reflects a change in one variable.

The covariance value can range from -âˆž to +âˆž, with a negative value indicating a negative relationship and a positive value indicating a positive relationship.

The greater this number, the more reliant the relationship. Positive covariance denotes a direct relationship and is represented by a positive number.

A negative number, on the other hand, denotes negative covariance, which indicates an inverse relationship between the two variables. Covariance is great for defining the type of relationship, but it's terrible for interpreting the magnitude.

`Types of Covariance`
Covariance can be either positive or negative. It is classified into two sorts based on this:

1. Positive Covariance
A positive covariance between two variables indicates that they are heading in the same direction. The variables, in this case, behave similarly. That is, if the values of one variable (more or smaller) correspond to the values of another variable, they are said to be in positive covariance.

2. Negative Covariance
When two variables have a negative covariance, the variables shift in the opposite direction. It is the inverse of positive covariance, in which higher values of one variable correlate to lower values of another and vice versa.

`Application Of Covariance`
The following are the most common applications of Covariance:

1. Simulating systems with multiple correlated variables is done using Cholesky decomposition. A covariance matrix helps determine the Cholesky decomposition because it is positive semi-definite. The matrix is decomposed by the product of the lower matrix and its transpose. 
2. To reduce the dimensions of large data sets, principal component analysis is used. To perform principal component analysis, an eigen decomposition is applied to the covariance matrix

`Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.`

Ans 14: Outliers affect the mean value of the data but have little effect on the median or mode of a given set of data.

```python
#Example 1: Without outliers
data = [2,2,3,3,3,4,5,6,7,8,9]
print("The mean is:",np.mean(data))
print("The median is:",np.median(data))
print("The mode is:",stats.mode(data))
```

```python
#Example 1: With outliers
data = [2,2,3,3,3,4,5,6,7,8,9,100]
print("The mean is:",np.mean(data))
print("The median is:",np.median(data))
print("The mode is:",stats.mode(data))
```

The above example 1 demonstrates how adding an outlier drastically changes the mean whereas it has very little effect on median and no effect on mode.
