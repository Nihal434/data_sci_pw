`Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.`

```python
# Ans 2:
# H0: Distribution of colors of M&Ms matches the expected distribution
# Ha: Distribution of colors of M&Ms does not match the expected distribution
import scipy.stats as stat
import numpy as np
#Expected distribution: 20% blue, 20% orange, 20% green, 10% yellow, 10% red , 20% brown
#Let's take a random sample of 100: 30 blue, 15 orange, 25 green, 15 yellow, 5 red, 10 brown\
observed = [30,15,25,15,5,10]
expected = [20,20,20,10,10,20] #based on expected proportions and a sample of 100

sum(observed),sum(expected)
```

```python
#performing chi-square goodness of fit test
chi_stat, p_value = stat.chisquare(observed,expected)
print(chi_stat,p_value)
```

```python
#calculating critical value
significance = 0.05
dof = len(observed) - 1
critical_value = stat.chi2.ppf(1-significance,dof) #critical limit for null hypothesis
print(critical_value)
if chi_stat > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

`Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

#x-axis ranges from 0 to 20 with .001 steps
x = np.arange(0, 20, 0.005)
y = chi2.pdf(x, df=10)

#plot Chi-square distribution with 4 degrees of freedom
plt.plot(x, y,color="blue",label='df: 10')


#labels
plt.xlabel("X")
plt.ylabel("Density")
plt.title("Chi square distribution", fontsize =14)

#the significane for chi statistic (dof=10) of 15 = 0.1 or 10%
#the area of curve falling under 10% corresponds to x > 18

plt.fill_between(x=x,y1=y,where=(x>18),color='red',label="chi statistic : 15 at dof = 10")

#add legend
plt.legend()
```

`Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.`

```python
#H0:The observed frequencies match the expected expected frequencies
#Ha:The observed frequencies are different from the expected expected frequencies
## for a fair coin the probablity of head = probablity of tails = 0.5, Therefore for 100 tries , Expected heads = tails = 50
## Observed heads = 100 - 45 = 55 and Tails = 45
import scipy.stats as stat
import numpy as np
observed = [55,45]
expected = [50,50]
sum(observed),sum(expected)
```

```python
## Chi - square goodness of fit

chisqaure_test_statistics, p_value = stat.chisquare(observed,expected)
print(chisqaure_test_statistics,p_value)
```

```python
## find the critical value
significance = 0.05
dof = len(observed) -1
critical_value = stat.chi2.ppf(1-significance,dof)
print(critical_value)
if chisqaure_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

`Inference`: The observed frequencies match the expected expected frequencies

`Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.`

||Lung Cancer: Yes|Lung Cancer: No|
|-|---------------|---------------|
|Smoker |60 |140|
|Non-smoker |30 |170|

`Use a significance level of 0.05.`


```python
#H0: There is no association between lung cancer and smoking
#Ha: There is a significant association between lung cancer and smoking
#significance assumed = 0.05
significance = 0.05

import pandas as pd
import scipy.stats as stat
from scipy.stats import chi2_contingency

df = pd.DataFrame({"Lung Cancer: Yes":[60,30],"Lung Cancer: No":[140,170]},index=["Smoker","Non-smoker"])
c, p, dof, expected = chi2_contingency(df)
print("The chi-sqaure statistic is:",c,"\n","The p-value is:",p)
```

```python
#calculating critical value
significance = 0.05
critical_value = stat.chi2.ppf(1-significance,dof) #critical limit for null hypothesis
print("The critical value is:",critical_value)
print("The chi-sqaure statistic is:",c)
if c > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

`Inference`: There is a significant association between lung cancer and smoking

`Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.`

Since the population variance is unknown and sample size not > 30 we use T-test in this test case to estimate the population mean is significantly different from 70
H0:μ = 70
Ha:μ != 70 (2 -tailed)

Data given: X̄ = 72, μ = 70, S = 10 , n = 30 , dof = 30-1 = 29

Decision rule: if the t value is below -2.045 or above +2.045 the H0 will be rejected

calcualting t statistic:

T = (X̄ – μ) / S/√n

T = (72 - 70) / 10/√30 = 1.0954

Since T (1.0954) is less than 2.045 the test fails to reject the null hypothesis

`Inference`: The population mean is not significantly different from 70
